```python
import os
import pandas as pd
import numpy as np

PATH="Dataset/diabetes.csv"

def load_data(path=PATH):
    col_names=['pregnant', 'glucose', 'bp', 'skin', 'insulin', 'bmi', 'pedigree', 'age', 'label']
    return pd.read_csv(PATH, header=None,  names=col_names)


```

```python
diabetes=load_data()
diabetes.head()
```

```python
#clean data optional

diabetes[['pregnant', 'glucose', 'bp', 'skin', 'insulin']] = diabetes[['pregnant', 'glucose', 'bp', 'skin', 'insulin']].replace(0, np.NaN)
# replace  with miss
diabetes.fillna(diabetes.mean(), inplace=True)
# summarize the number of rows and columns in the dataset
diabetes.head()
```

```python
#split dataset in features and target variable
#"pregnant","glucose","bp","skin","insulin","bmi","pedigree","age","label"
feature_cols = ["pregnant","glucose","bp","skin","insulin","bmi","pedigree","age"]
X = diabetes[feature_cols] # Features
y = diabetes.label # Target variable
```

```python
#split the dataset into training and testing
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test=train_test_split(X,y, test_size=0.2, random_state=0)
```

```python
means = np.mean(X_train, axis=0)
stds = np.std(X_train, axis=0)
X_train = (X_train - means)/stds
X_test = (X_test - means)/stds
```

```python
logreg = LogisticRegression(max_iter=1000)
logreg.fit(X_train,y_train)
```

```python
#pred=perceptron.predict(x_test)
y_pred=logreg.predict(X_test)
```

```python
print('Accuracy: %.4f' % accuracy_score(y_test, y_pred))
```

```python
#output should be 1
test=[[6, 148,72, 35, 155, 33, 0.63, 50]]
y_pred=logreg.predict(test)
print(y_pred)
```

```python
coeff = list(logreg.coef_[0])
labels = list(X_train.columns)
features = pd.DataFrame()
features['Features'] = labels
features['importance'] = coeff
features.sort_values(by=['importance'], ascending=True, inplace=True)
features['positive'] = features['importance'] > 0
features.set_index('Features', inplace=True)
features.importance.plot(kind='barh', figsize=(11, 6),color = features.positive.map({True: 'blue', False: 'red'}))
plt.xlabel('Importance')
```
